```python
weather_data['Date/Time'] = pd.to_datetime(weather_data['Date/Time'])
weather_data
```

```python
weather_data['month'] = weather_data['Date/Time'].dt.month
weather_data
```

```python
sns.set_palette('viridis')
sns.displot(weather_data['Temp_C'],kde = True)
plt.show()
```

```python
plt.figure(figsize = (20,10))
plt.plot(weather_data['Date/Time'],weather_data['Temp_C'], marker = 'o', color = 'indigo')
plt.xlabel('Date/Time')
plt.ylabel('Temperature Variation')
plt.show()
```

```python
sns.boxplot(x = 'month', y = 'Rel Hum_%', data = weather_data, palette ='pastel',)

plt.show()
```

```python
sns.stripplot(x = 'month', y = 'Wind Speed_km/h', data = weather_data, palette = 'dark')
plt.show()
```

```python
sns.pointplot(x = 'month', y = 'Visibility_km', data = weather_data)
plt.show()
```

```python
print(weather_data.columns)

column_name = 'Weather'
if column_name in weather_data.columns:
    weather_data_encoded = pd.get_dummies(weather_data, columns=[column_name])
    correlation_matrix = weather_data_encoded.corr()
    print(correlation_matrix)
else:
    print(f"Column '{column_name}' not found in the DataFrame.")
```

```python
plt.figure(figsize = (10,9))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5, annot_kws={"size": 8}, square=True)
plt.show()
```

```python
sns.lineplot(x = 'month', y = 'Press_kPa', data = weather_data)
plt.xlabel('Month')
plt.ylabel('Press_kPa')
plt.title('Atmospheric Pressure Over Time')
plt.show()
```
